```python
import numpy as np
import matplotlib.pyplot as plt


categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))


plt.bar(categories, values, color=['blue', 'green', 'red', 'orange', 'purple'])


plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Random Data Bar Plot")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


categories = ['A', 'B', 'C', 'D', 'E']
num_bars = 3  
values = np.random.randint(10, 100, size=(num_bars, len(categories)))

bar_width = 0.2
x = np.arange(len(categories))

colors = ['red', 'blue', 'green']
labels = ['Set 1', 'Set 2', 'Set 3']

for i in range(num_bars):
    plt.bar(x + i * bar_width, values[i], width=bar_width, color=colors[i], label=labels[i])


plt.xticks(x + bar_width, categories)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Multiple Bar Chart Example")
plt.legend()


plt.show()
```
